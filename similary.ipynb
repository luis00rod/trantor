```python
df_corpus = df[["name_cleaned", "target_phrase"]].rename(columns = \
                                    {"name_cleaned": "sentence1", "target_phrase": "sentence2"})
df_corpus
```

```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('multi-qa-MiniLM-L6-cos-v1')

query_embedding = model.encode('How big is London')
passage_embedding = model.encode(['London has 9,787,426 inhabitants at the 2011 census',
                                  'London is known for its finacial district'])

print("Similarity:", util.dot_score(query_embedding, passage_embedding))
```

```python
model = SentenceTransformer('bert-base-nli-mean-tokens')
```

```python
df_corpus.to_csv('data_embedding.csv')
```

```python
df_corpus["sentence_numeric1"] = df_corpus["sentence1"].apply(lambda row: model.encode(row))
df_corpus["sentence_numeric2"] = df_corpus["sentence2"].apply(lambda row: model.encode(row))

df_corpus
```

```python
df_corpus["sentence_numeric1"][0].shape
```

```python
def cos_sim_tuned(row) -> float:
    result = util.cos_sim(row.sentence_numeric1, row.sentence_numeric2).numpy()
    return result[0][0]
```

```python
df_corpus["cos_similiraty"] = df_corpus.apply(lambda row: cos_sim_tuned(row), axis = 1)
df_corpus
```
